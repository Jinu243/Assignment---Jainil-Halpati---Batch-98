1.Plot Total Sales Per Month for Year 2011. How the total sales have increased over months in Year 2011. Which month has lowest Sales?

```python
import numpy as np
import pandas as pd
import os 
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20,10)
```

```python
df = pd.read_csv('BigMartSalesData.csv',delimiter=',')
```

```python
df.head()
```

```python
df.info()
```

Sales for year 2011 by month


```python
sales_2011 = df[df['Year'] == 2011]
```

```python
sales_2011
```

```python
sales_2011_by_month = sales_2011.groupby('Month').sum()['Amount']
```

```python
print(sales_2011_by_month)
```

Plot the graph

```python
plt.plot(sales_2011_by_month.index,sales_2011_by_month.values)

plt.xlabel('Sales in Euro')
plt.ylabel('Month Number')
plt.title('Sales Per Month in Year 2011')
plt.show()
```

Save the plot Locally

```python
plt.savefig('Year2011MonthlySales')
```

2.Plot Total Sales Per Month for Year 2011 as Bar Chart. Is Bar Chart Better to visualize than Simple Plot?

```python
plt.bar(sales_2011_by_month.index,sales_2011_by_month.values,color="C2") # Change the color and play
#enhancements in code :- 1.Change the bar chart to show the value of bar
for i, value in enumerate(sales_2011_by_month.values):
    plt.text(sales_2011_by_month.index[i], value, str(round(value, 2)), ha='center', va='bottom')
    
plt.xlabel('Sales in Euro')
plt.ylabel('Month Number')
plt.title('Sales Per Month in Year 2011')
plt.show()
```

3.Plot Pie Chart for Year 2011 Country Wise. Which Country contributes highest towards sales?

```python
sales_country_wise = sales_2011.groupby('Country').sum()['Amount']

plt.title('Country Wise Contribution For 2011')
#enhancements in code :-2.In Pie Chart Play With Parameters shadow=True, startangle=90 and see how different the chart looks
plt.pie(sales_country_wise.values,labels=sales_country_wise.index,autopct='%1.1f%%',shadow=True,startangle=90,labeldistance=1.1,rotatelabels=True)
plt.show()
```
