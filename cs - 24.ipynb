Data Manipulation:

Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes.


```python
prisoners["Total_Benefited"]=prisoners.iloc[:,2:].sum(axis=1)
prisoners
```

Create a new Data Frame to Store column-wise total. Name New Data Frame as prisoners_total

```python
prisoners_total  =  prisoners
```

Create a new row - “Prisoners Total'” that is the sum of all inmates benefitted through each mode across all states.

```python
total_row = prisoners.iloc[:,2:].sum()
```

```python
total_row['STATE/UT'] = 'Prisoners Total'
total_row['YEAR'] = 'All Years'
```

```python
total_row = pd.DataFrame([total_row])

prisoners_total = pd.concat([prisoners, total_row], ignore_index=True)
```

```python
prisoners_total
```

Change the data type of the year column to String from integer. Since there is no addition for the YEAR column, NaN will be populated to this column and the data type changes from integer to float automatically. To avoid this explicitly change the data type to string.

```python
prisoners_total['YEAR'] = prisoners_total['YEAR'].astype(str)
```

Add column-wise total where the data type is numeric

```python
prisoners_total.loc['Prisoners Total'] = prisoners_total.iloc[:,2:].sum(numeric_only=True)
```

```python
prisoners_total
```

Fist two columns will be populated as NaN , Replace the NaN with empty string for better formatting 

```python
prisoners_total.loc[prisoners_total.index[-1], ['STATE/UT','YEAR']] = ''
prisoners_total
```

Plotting:

Make a bar plot with each state name on the x -axis and their total benefitted inmates as their bar heights. Which state has the maximum number of beneficiaries?


```python
xlabels = prisoners['STATE/UT'].values
plt.figure(figsize=(20, 10))

plt.xticks(np.arange(xlabels.shape[0]),xlabels,rotation='vertical', fontsize=18)
plt.xticks

plt.bar(np.arange(prisoners.iloc[:-1].values.shape[0]),prisoners['Total_Benefited'].iloc[:-1],align='edge')
plt.show()
```

Make a pie chart that depicts the ratio among different modes of benefits.

```python
labels = prisoners.columns[2:-1]
sizes = prisoners_total.iloc[-1][2:-1].values
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```
