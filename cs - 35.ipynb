```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import LinearSVC,SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("College.csv")
df.head()
```

```python
X, y = df.iloc[:, 1:].values, df.iloc[:, 0].values
target_encoder = LabelEncoder()
y = target_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print(X_train.shape)
```

```python
classifier = LinearSVC()


classifier.fit(X_train,y_train)

y_predict = classifier.predict(X_test)

print (classifier.score(X_test,y_test))
```

```python
print(confusion_matrix(y_predict,y_test))
```

```python
accuracy = accuracy_score(y_predict,y_test)
print (accuracy )
```

```python
print(classification_report(y_test, y_predict))
```

```python
scaler = StandardScaler()
X, y = df.iloc[:, 1:].values, df.iloc[:, 0].values
X = scaler.fit_transform(X)
```

```python
target_encoder = LabelEncoder()
y = target_encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print(X_train.shape)
```

```python
classifier = SVC()
classifier.fit(X_train,y_train)
classifier.score(X_test,y_test)
```

```python
y_predict = classifier.predict(X_test)
print(confusion_matrix(y_predict,y_test))
```

```python
accuracy = accuracy_score(y_predict,y_test)
print (accuracy )
```

```python
print(classification_report(y_test, y_predict))
```

```python
C_range = np.logspace(-2, 10, 13)
gamma_range = np.logspace(-9, 3, 13)
param_grid = dict( gamma=gamma_range,C=C_range)
grid = GridSearchCV(SVC(), param_grid=param_grid,verbose=2)
grid.fit(X_train, y_train)
```

```python
print("The best parameters are %s with a score of %0.2f"
      % (grid.best_params_, grid.best_score_))
```

```python
y_predict = grid.predict(X_test)
print(confusion_matrix(y_predict,y_test))
```

```python
accuracy = accuracy_score(y_predict,y_test)
print (accuracy )
```

```python
print(classification_report(y_test, y_predict))
```

```python
params = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001]}
print (params)
rgrid =  RandomizedSearchCV(SVC(), param_distributions=params, n_iter= 12,cv = 10,  verbose=2 )

rgrid.fit(X_train, y_train)
```

```python
print("The best parameters are %s with a score of %0.2f"
      % (rgrid.best_params_, rgrid.best_score_))
```

```python
y_predict = rgrid.predict(X_test)
print(confusion_matrix(y_predict,y_test))
```

```python
accuracy = accuracy_score(y_predict,y_test)
print (accuracy )
```

```python
print(classification_report(y_test, y_predict))
```
