```python
x = df.iloc[:,[1,2]].values
print (x)
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans (n_clusters =  i , init = 'k-means++')
    kmeans.fit(x)
    wcss.append (kmeans.inertia_)
display (wcss)
```

```python
plt.plot (range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel ('wcss')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
df_analyze = df.drop('id',axis=1)
kmeans.fit(df_analyze)
```

```python
display (kmeans.cluster_centers_)
```

```python
print (kmeans.labels_)
print (len(kmeans.labels_))
```

```python
print (type(kmeans.labels_))
unique, counts = np.unique(kmeans.labels_, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
df_analyze['cluster'] = kmeans.labels_
sns.set_style('whitegrid')
#aspect (the ratio of the width to the height) 
# palette = colour 
sns.lmplot(data=df_analyze,x= 'mean_dist_day',y= 'mean_over_speed_perc', hue='cluster',
           palette=['blue','green'],aspect=2,fit_reg=False)
plt.show()
```

```python
Final = pd.concat([df,pd.DataFrame(kmeans.labels_, columns = ['Cluster'])],axis =1)
Final.to_excel("D:\\top mentor assignment\\Batch 98 Day 57\\ML Case Studies with Answer\\CS 37 Ans -Unsupervised Learning-1\\driver-data_2_Cluster.xlsx")
display (Final)
```

```python
kmeans =  KMeans (n_clusters =  4 , init = 'k-means++')
kmeans = kmeans.fit(x)
y_kmeans = kmeans.predict(x)
print (y_kmeans)
```

```python
unique, counts = np.unique(kmeans.labels_, return_counts=True)
print(dict(zip(unique, counts)))
plt.scatter (x[y_kmeans == 0,0],x[y_kmeans == 0,1] ,s = 100, c='r', label = 'Cluster 1')
plt.scatter (x[y_kmeans == 1,0],x[y_kmeans == 1,1] ,s = 100, c='b', label = 'Cluster 2')
plt.scatter (x[y_kmeans == 2,0],x[y_kmeans == 2,1] ,s = 100, c='g', label = 'Cluster 3')
plt.scatter (x[y_kmeans == 3,0],x[y_kmeans == 3,1] ,s = 100, c='c', label = 'Cluster 4')
plt.legend()
plt.show()
```

```python
Final = pd.concat([df,pd.DataFrame(y_kmeans, columns = ['Cluster'])],axis =1)
Final.to_excel("D:\\top mentor assignment\\Batch 98 Day 57\\ML Case Studies with Answer\\CS 37 Ans -Unsupervised Learning-1\\driver-data_4_Cluster.xlsx")
display(Final)
```
